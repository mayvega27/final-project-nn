#**Arquitectura/Métodos**

## Arquitectura

Arquitectura de tipo auto-encoder con tres terminos: 

1. Un codificador que actúa como una red neuronal convolucional de tipo ResNet que disminuye las dimensiones y aumenta las características.
2. Un decodificador con la misma arquitectura del codificador, pero haciendo el proceso inverso. Combina pasos intermedios del codificador con la salida final del codificador para reconstruir las regiones tumorales.
3. Decodificador VAE. Toma la salida del codificador y samplea muestras de la distribución gausiana para reconstruir la imagen original.

El decodificador VAE tiene como propósito ayudar a regularizar el proceso del codificador a medida que extrae características.



Encoder.png



Imagen tomada de [3d-mri-brain-tumor-segmentation-using-autoencoder-regularization](https://github.com/IAmSuyogJadhav/3d-mri-brain-tumor-segmentation-using-autoencoder-regularization) por Andriy Myronenko, NVIDIA.

## Loss Function

La loss function se contruyo a partir de tres losses.

1. Loss function asociada al decodificador: mide como de bien las predicciones se parecen a las etiquetas reales.

$$L_{dice} = \frac{2 \sum p_{true} \times p_{pred}}{\sum p^2_{true} + \sum p^2_{pred} + ɛ}$$

2. Loss functions asociadas al decodificador VAE: $L_{L2}$ mide que tanto se parece la imagen reconstruida a la original bajo la metrica euclidiana. $L_{KL}$ mide que tanto se parece la imagen reconstruida a la original bajo sus distribuciones de probabilidad.

$$L_{L2} = \|{I_{input} - I_{pred}}\|_2^2 $$

$$L_{KL} = \frac{1}{N}\Sigma\mu^2 + \sigma^2 - \log\sigma^2 - 1$$

Loss function final:

$$L = L_{dice} + 0.1 L_{L2} + 0.1 L_{KL}$$

Los coeficientes de las loss functions del termino VAE se encontraron mediante una búsqueda de hyperparametros que balanceara el peso del decodificador y el decodificador VAE.

```python
import sys
sys.path.append('/content/drive/MyDrive/Brats/tumor-segmentation-master/notebooks/colab_notebooks')

import SimpleITK as sitk  # For loading the dataset
import numpy as np  # For data manipulation
from model import build_model  # For creating the model
import glob  # For populating the list of files
from scipy.ndimage import zoom  # For resizing
import re  # For parsing the filenames (to know their modality)
import os

from keras.callbacks import History 
history = History()
```

```python
def resize(img, shape, mode='constant', orig_shape=(155, 240, 240), order=3):
    """
    Wrapper for scipy.ndimage.zoom suited for MRI images.
    """
    assert len(shape) == 3, "Can not have more than 3 dimensions"
    factors = (
        shape[0]/orig_shape[0],
        shape[1]/orig_shape[1], 
        shape[2]/orig_shape[2]
    )

    # Resize to the given shape
    return zoom(img, factors, mode=mode, order=order)


def preprocess(img, out_shape=None):
    """
    Preprocess the image.
    Just an example, you can add more preprocessing steps if you wish to.
    """
    if out_shape is not None:
        img = resize(img, out_shape, mode='constant')
    
    # Normalize the image
    mean = img.mean()
    std = img.std()
    return (img - mean) / std


def preprocess_label(img, out_shape=None, mode='nearest'):
    """
    Separates out the 3 labels from the segmentation provided, namely:
    GD-enhancing tumor (ET — label 4), the peritumoral edema (ED — label 2))
    and the necrotic and non-enhancing tumor core (NCR/NET — label 1)
    """
    ncr = img == 1  # Necrotic and Non-Enhancing Tumor (NCR/NET)
    ed = img == 2  # Peritumoral Edema (ED)
    et = img == 4  # GD-enhancing Tumor (ET)
    print(ncr)
    
    if out_shape is not None:
        ncr = resize(ncr, out_shape, mode=mode)
        ed = resize(ed, out_shape, mode=mode)
        et = resize(et, out_shape, mode=mode)

    return np.array([ncr, ed, et], dtype=np.uint8)

def unprocess_label(img, out_shape=None, mode='nearest'):
    """
    Recombines the 3 labels to the segmentation format from brats.
    """
    ncr = img[0]
    ed = img[1]
    et = img[2]
    # ncr = img[0,:,:,:] # Necrotic and Non-Enhancing Tumor (NCR/NET)
    # ed = img[1,:,:,:]  # Peritumoral Edema (ED)
    # et = img[2,:,:,:]  # GD-enhancing Tumor (ET)
    print(np.unique(ncr))
    print(np.unique(ed))
    print(np.unique(et))
    
    if out_shape is not None:
        ncr = resize(ncr, out_shape, mode=mode, orig_shape=(80, 96, 64), order=0)
        ed = resize(ed, out_shape, mode=mode, orig_shape=(80, 96, 64), order=0)
        et = resize(et, out_shape, mode=mode, orig_shape=(80, 96, 64), order=0)

    # update label values
    # no update needed for ncr
    ed = np.where(ed == 1,2,0)
    et = np.where(et == 1,4,0)

    out = ncr + ed + et

    print(np.unique(ncr))
    print(np.unique(ed))
    print(np.unique(et))
    
    print('out')
    print(out.shape)
    print(np.unique(out))
    # return np.array([ncr, ed, et], dtype=np.uint8)
    return out
    
```

```python
def read_img(img_path):
    """
    Reads a .nii.gz image and returns as a numpy array.
    """
    return sitk.GetArrayFromImage(sitk.ReadImage(img_path))
```

```python
import tensorflow as tf
import keras.backend as K
from sklearn.utils.extmath import cartesian
```

```python
def dice_coef(y_true, y_pred):
    smooth = 1
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def dice_coef_loss(y_true, y_pred):
    return -dice_coef(y_true, y_pred)
```

La misma imagen después de redimensionar a (80, 96, 64)

```python
img = preprocess_label(read_img(seg[3]), out_shape=(80, 96, 64), mode='nearest')[1][50]
print(img.shape)
print(np.unique(img))
print(img.sum())
plt.imshow(img, cmap='Greys_r')
```

Este fenómeno de "desvanecimiento de pixeles" es un potencial problema en la implementación del modelo. De manera que, aunque se intuye que es inevitable el redimensionar algunas máscaras durante el proceso de entrenamiento, debemos evitar redimensionar en exceso.

A continuación se ejemplifica una técnica para reducir este tipo de ruido.

```python
import cv2
```

- Original Image > preprocess_label > Morphological Closing



```python
kernel = np.ones((3, 3))
img_closed = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel, iterations=3)
print(np.unique(img_closed))
print(img_closed.sum())
plt.imshow(img_closed, cmap='Greys_r')
```

- Original Image > preprocess_label > Morphological Dilation

```python
kernel = np.ones((3, 3))
img_dilated = cv2.dilate(img, kernel, iterations=1)
print(np.unique(img_dilated))
print(img_dilated.sum())
plt.imshow(img_dilated, cmap='Greys_r')
```

```python
def show_slices(slices):
    """ Function to display row of image slices """
    fig, axes = plt.subplots(1, len(slices),figsize=(20,10))
    for i, slice in enumerate(slices):
        # axes[i].imshow(slice.T, cmap="gray", origin="lower")
        axes[i].imshow(slice.T, origin="lower")

def process_slices(data_frame,mode=3):
    epi_img_data = data_frame[mode]

    slice_0 = epi_img_data[40, :, :]
    slice_1 = epi_img_data[:, 48, :]
    slice_2 = epi_img_data[:, :, 32]

    
    show_slices([slice_0, slice_1, slice_2])
```
